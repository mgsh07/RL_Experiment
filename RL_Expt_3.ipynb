```python
import gym, gym_bandits, numpy as np, warnings, matplotlib.pyplot as plt
env = gym.make('MultiarmedBandits-v0')
print(env.observation_space)
print(env.action_space)
```

```python
n_arms, num_epiepisodessodes, epsilon, all_rewards, action_history = env.action_space.n, 500, 0.1, [], []
q_values, counts, reward_sums = np.zeros(n_arms), np.zeros(n_arms), np.zeros(n_arms)
warnings.filterwarnings('ignore', category = DeprecationWarning)
```

```python
for episode in range(num_episodes):
  env.reset()
  if np.random.rand() < epsilon:
    action = env.action_space.sample()
  else:
    action = np.argmax(q_values)
  observation, reward, done, info = env.step(action)
  counts[action] += 1
  reward_sums[action] += reward
  q_values[action] = reward_sums[action] / counts[action]
  all_rewards.append(reward)
  action_history.append(action)
  env.close()
```

```python
print(f"Estimated Q-values for each arm: {np.round(q_values, 2)}")
print(f"Number of times each arm was selected: {counts.astype(int)}")
print(f"Average reward over all episodes: {round(np.mean(all_rewards), 3)}")

plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(np.cumsum(all_rewards), color='blue')
plt.title("Cumulative Reward Over Episodes")
plt.xlabel("Episode")
plt.ylabel("Cumulative Reward")
plt.subplot(1, 2, 2)
plt.bar(np.arange(n_arms), counts, color='green')
plt.title("Action Selection Frequency")
plt.xlabel("Arm Index")
plt.ylabel("Times Selected")
plt.tight_layout()
plt.show()
```
